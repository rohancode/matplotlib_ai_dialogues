```python
# What was the birth rate per 1000 people in India from 2010 to 2019
```

```python
from pandas_datareader import wb
import datetime

start = 2010
end = 2019
df = wb.download(indicator='SP.DYN.CBRT.IN', country=['IN'], start=start, end=end)
print(df)
```

```python
# Draw me a chart for this birth rate count
```

```python
from pandas_datareader import wb
import datetime
import pylab as plt

start = 2010
end = 2019
df = wb.download(indicator='SP.DYN.CBRT.IN', country=['IN'], start=start, end=end)
df = df.reset_index()
fig, ax = plt.subplots()
ax.plot(df.year, df.iloc[:,-1])
```

```python
# Add grid line on this chart
```

```python
from pandas_datareader import wb
import datetime
import pylab as plt

start = 2010
end = 2019
df = wb.download(indicator='SP.DYN.CBRT.IN', country=['IN'], start=start, end=end)
df = df.reset_index()
fig, ax = plt.subplots()
ax.plot(df.year, df.iloc[:,-1])
ax.grid(True)
```
