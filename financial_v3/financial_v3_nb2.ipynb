```python
# Let's rebuild this dataframe with only Consumer Price Index relevant data and remove everything else
```

```python
import pandas_datareader.data as web
import datetime
import pandas as pd

df = web.DataReader('ticker=RGDPUS', 'econdb')
df = pd.DataFrame({
    "date": df.iloc[:,0].index,
    "value": df.iloc[:,0].values
})
print(df)
```

```python
# And make a chart from the processed data
```

```python
import pandas_datareader.data as web
import datetime
import pandas as pd
import pylab as plt

df = web.DataReader('ticker=RGDPUS', 'econdb')
df = pd.DataFrame({
    "date": df.iloc[:,0].index,
    "value": df.iloc[:,0].values
})
fig, ax = plt.subplots()
ax.plot(df.date, df.value)
```

```python
import pandas_datareader.data as web
import datetime
import pandas as pd
import pylab as plt

df = web.DataReader('ticker=RGDPUS', 'econdb')
df = pd.DataFrame({
    "date": df.iloc[:,0].index,
    "value": df.iloc[:,0].values
})
fig, ax = plt.subplots(figsize=(16,5))
ax.plot(df.date, df.value)
```
