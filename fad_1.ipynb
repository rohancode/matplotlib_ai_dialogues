```python
import pandas_datareader.data as web
import datetime
import pylab as plt 

start = datetime.datetime(2008, 1, 1)
end = datetime.datetime(2008, 12, 1)
data = web.DataReader('GDP', 'fred', start, end)

fig, ax = plt.subplots()
ax.plot(data.index, data.GDP)
```
