```python
# What is the trend of "High" over the range?
```

```python
import pandas_datareader.data as web
import datetime
import pylab as plt 

data = web.DataReader("AAPL", 'yahoo', datetime.date.today() - datetime.timedelta(days=180))
data = data[['High']]

fig, ax = plt.subplots()
ax.plot(data.index, data.High)
```
